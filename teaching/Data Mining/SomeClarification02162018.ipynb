```python
x = np.linspace(-4,10,100)   
f1 = (3/5) * (1/np.sqrt(2*np.pi)) * np.exp(-(x-2)**2/2) + (2/5) * (1/np.sqrt(2*np.pi)) * np.exp(-(x-4)**2/2)
f2 = (3/5) * (1/np.sqrt(2*np.pi)) * np.exp(-(x-4)**2/2) + (2/5) * (1/np.sqrt(2*np.pi)) * np.exp(-(x-5)**2/2)
```

```python
# If given multiple x and y in plot(), except the color and linestyle you specify in the ''
# the other parameters would apply to all the lines
plot(x, f1, 'r-', x, f2, '--', linewidth = 5);

# In this case, if you want the legend, you might need the following setting
legend(['f1','f2']);
```

```python
# color and linstyle are overwritten

plot(x, f1, 'r-', x, f2, '--', linewidth = 5, color = 'g', linestyle='dashed');
legend(['f1','f2']);
```
