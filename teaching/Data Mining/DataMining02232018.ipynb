### Whitening

- Preprocess data -- frequently used in signal processing
- Make features less redundant
> Less correlated with each other
>
> Same variance

#### Singular-value decomposition

> The generalization of the eigendecomposition 
>
> For example, a symmetric $ n \times n $ matrix with positive eigenvalues to any $ n \times m $ matrix
>
> $ X_{n \times m} = U W V^T$
- $ U_{n \times n} $, $ U^TU=I $
- $ W_{n \times m} $, diagonal
- $ V_{m \times m} $, $ V^TV=I $

#### Procedure of PCA

- Data matrix 
> $ X = (x_{ij})_{n \times m} $


- Centering
> $X_{new} = X - mean(X)$

- Decomposition
> Original data matrix 
 - $ X_{new} = U W V^T$
>
> Sample covariance matrix 
 - $ C = \frac{1}{n-1}\ X_{new} X_{new}^T = \frac{1}{n-1}\  \sum_i x_i x_i^T $
 - $ C = E \Lambda E^T $
>
> Hence
 - $ C = \frac{1}{n-1}\  UWV^T\ VWU^T = \frac{1}{n-1}\ U W^2 U^T = E \Lambda E^T $
 - $ E = U $
 - $ \Lambda = \frac{1}{n-1}\  W^2 $

- Choose first $k$ components
> $ XX = E_k^TX $

#### Procedure of Whitening

> $ Z = \Lambda^{-1/2} E^T X $
>
> $ C_Z = \frac{1}{n-1} Z Z^T = \frac{1}{n-1} \left(\Lambda^{-1/2} E^T X \right) \left(\Lambda^{-1/2} E^T X \right)^T = \Lambda^{-1/2} E^T \left(\frac{1}{n-1} X X^T \right) E \Lambda^{-1/2} = \Lambda^{-1/2} E^T C E \Lambda^{-1/2} = \Lambda^{-1/2} E^T E \Lambda E^T E \Lambda^{-1/2} = I $

```python
from scipy.stats import norm 
from scipy.linalg import svd
```

```python
# generate 2D (column) vectors
np.random.seed(seed=2018);
N = norm.rvs(0,1,(2,50)); 
N[0,:] *= 2 
f = +pi/6   # rotate by 30 deg
R = array([[cos(f), -sin(f)],
           [sin(f),  cos(f)]]) 
X = R.dot(N)

figure(figsize=(5,5)); 
subplot(111,aspect='equal');
plot(X[0,:],X[1,:],'x',alpha=0.6);
```

```python
X -= mean(X, axis=1).reshape(X[:,1].size,1)
U, W, V = svd(X)
E = U
Lambda = W**2 / (X[0,:].size-1)
```

```python
# Whitening
Z = diag(1/sqrt(Lambda)).dot(E.T).dot(X)

# Sample covariance matrix for new Z
Z.dot(Z.T) / (Z.shape[1]-1)
```

```python
from sklearn.decomposition import PCA

# Choose first 2 PCs by setting [n_componts] = 2 and whitening the data by setting [whiten] = True
pca = PCA(n_components=2, whiten=True)
```

```python
Z = pca.fit_transform(X.T)
Z.T.dot(Z) / (Z.shape[0])
```
